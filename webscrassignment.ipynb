# web scraping

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get('https://en.wikipedia.org/wiki/Web_scraping')
page.status_code
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
soup.title.string
```

```python
ataglist = []
for link in soup.find_all('a'):
    print(link.get('href'))
    ataglist.append(link.get('href'))
    ataglist
    adataframe = pd.DataFrame(ataglist)
```

```python
paragraphlist = []
for pera in soup.find_all('p'):
    print(pera.get_text())
    paragraphlist.append(pera.get_text())
    paragraphlist
    pdataframe = pd.DataFrame(paragraphlist)
```

```python
h1list = []
for heading1 in soup.find_all('h1'):
    print(heading1.get_text())
    h1list.append(heading1.get_text())
    h1list
    h1dataframe = pd.DataFrame(h1list)
```

```python
h2list = []
for heading2 in soup.find_all('h2'):
    print(heading2.get_text())
    h2list.append(heading2.get_text())
    h2list
    h2dataframe = pd.DataFrame(h2list)
```

```python
h3list = []
for heading3 in soup.find_all('h3'):
    print(heading3.get_text())
    h3list.append(heading3.get_text())
    h3list
    h3dataframe = pd.DataFrame(h3list)
```

```python
divlist = []
for allDives in soup.find_all('div'):
    print(allDives.get_text())
    divlist.append(allDives.get_text())
    divlist
    divdataframe = pd.DataFrame(divlist)
```

```python
listsOfList = []
for lists in soup.find_all('li'):
    print(lists.get_text())
    listsOfList.append(lists.get_text())
    lidataframe = pd.DataFrame(listsOfList)
```

```python
all_tags = {"All_h1_tags": [h1dataframe],"All_h2_tags":[h2dataframe],"All_h3_tags":[h3dataframe],"AnchorTags":[adataframe],"All_P_tags":[pdataframe],"All_Div_tags":[divdataframe],"All_Li_tags":[lidataframe]}
```

```python
all_tags
```

```python
all_data_file = pd.DataFrame(all_tags)
```

```python
all_data_file.to_csv("allScrappeddata.csv")
```
